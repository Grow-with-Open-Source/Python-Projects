```python
# =====================================
# 💳 DBSCAN Clustering on Credit Card Fraud Dataset
# =====================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from google.colab import files
import io
from IPython.display import display

# Step 1️⃣: Upload CSV file
uploaded = files.upload()
csv_filename = list(uploaded.keys())[0]

# Step 2️⃣: Load dataset
df = pd.read_csv(io.BytesIO(uploaded[csv_filename]))
print("✅ Credit Card Dataset Loaded Successfully!\n")
display(df.head())

# Step 3️⃣: Clean column names
df.columns = df.columns.str.strip()
print("\n📋 Columns in dataset:")
print(df.columns.tolist())

# Step 4️⃣: Separate target (for reference only, NOT used in clustering)
if 'Class' in df.columns:
    y_true = df['Class']   # 1 = Fraud, 0 = Normal
    df = df.drop(columns=['Class'])

# Step 5️⃣: Select numerical features
df_clean = df.select_dtypes(include=[np.number])

# Handle missing or infinite values
df_clean = df_clean.replace([np.inf, -np.inf], np.nan).dropna()

# Step 6️⃣: Standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clean)

# Step 7️⃣: Apply DBSCAN
dbscan = DBSCAN(
    eps=1.5,        # Tunable parameter
    min_samples=5
)
clusters = dbscan.fit_predict(X_scaled)

# Step 8️⃣: Add cluster labels
df_clean['Cluster'] = clusters
display(df_clean.head(10))

# Step 9️⃣: Cluster distribution
print("\n📊 Cluster Distribution:")
print(df_clean['Cluster'].value_counts())

# Step 🔟: Silhouette Score (excluding noise points)
mask = clusters != -1
unique_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)

if unique_clusters > 1:
    sil_score = silhouette_score(X_scaled[mask], clusters[mask])
    print(f"\n🧩 Silhouette Score (excluding noise): {sil_score:.4f}")
else:
    print("\n⚠️ Not enough clusters for silhouette evaluation.")

# Step 1️⃣1️⃣: PCA Visualization (Sampling for performance)
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# Create DataFrame for plotting
plot_df = pd.DataFrame({
    'PC1': X_pca[:, 0],
    'PC2': X_pca[:, 1],
    'Cluster': clusters
})

# Sample data to avoid rendering issues
plot_df_sample = plot_df.sample(n=10000, random_state=42)

plt.figure(figsize=(8, 6))
plt.scatter(
    plot_df_sample['PC1'],
    plot_df_sample['PC2'],
    c=plot_df_sample['Cluster'],
    cmap='tab10',
    alpha=0.6
)

plt.title("💳 DBSCAN Clustering on Credit Card Transactions (PCA Sample)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.colorbar(label="Cluster")
plt.show()
```
