```python
# =====================================
# 🧬 DBSCAN Clustering on Breast Cancer Dataset (Unsupervised Learning)
# =====================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from google.colab import files
import io
from IPython.display import display

# Step 1️⃣: Upload CSV file
uploaded = files.upload()
csv_filename = list(uploaded.keys())[0]

# Step 2️⃣: Load dataset
df = pd.read_csv(io.BytesIO(uploaded[csv_filename]))
print("✅ File Loaded Successfully!\n")
display(df.head())

# Step 3️⃣: Clean column names
df.columns = df.columns.str.strip()
print("\n📋 Columns in dataset:")
print(df.columns.tolist())

# Step 4️⃣: Data Preprocessing
# Drop 'id' or unnamed columns
df = df.drop(columns=[col for col in df.columns if 'id' in col.lower() or 'unnamed' in col.lower()], errors='ignore')

# Convert 'diagnosis' to numeric for reference (not used for clustering)
if 'diagnosis' in df.columns:
    df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})

# Select only numeric columns
df_clean = df.select_dtypes(include=[np.number])

# Remove NaN or infinite values
df_clean = df_clean.replace([np.inf, -np.inf], np.nan).dropna()

# Step 5️⃣: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clean)

# Step 6️⃣: Apply DBSCAN Clustering
dbscan = DBSCAN(eps=2.5, min_samples=5)  # you can tune eps
clusters = dbscan.fit_predict(X_scaled)

# Step 7️⃣: Add cluster labels to DataFrame
df_clean['Cluster'] = clusters
display(df_clean.head(10))

# Step 8️⃣: Evaluate clustering quality (only if >1 cluster)
unique_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
if unique_clusters > 1:
    sil_score = silhouette_score(X_scaled, clusters)
    print(f"\n🧩 Silhouette Score: {sil_score:.4f} (higher = better clustering)")
else:
    print("\n⚠️ Only one cluster found — silhouette score not applicable.")

# Step 9️⃣: Show cluster distribution
print("\n📊 Cluster Counts:")
print(df_clean['Cluster'].value_counts())

# Step 🔟: Visualize Clusters (PCA 2D Projection)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters, palette='rainbow', legend='full')
plt.title("🌈 DBSCAN Clustering (PCA Visualization)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Cluster")
plt.show()
```
